### LOGISTICS REGRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder

cols_to_encode = ['job', 'marital', 'education', 'default', 'housing',
                   'loan', 'contact', 'month','poutcome','y']

le = LabelEncoder()

for x in cols_to_encode:
    bank_data[x] = le.fit_transform(bank_data[x])

features = bank_data[['age', 'job', 'marital', 'education', 'default', 'balance', 'housing',
       'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays',
       'previous', 'poutcome']]

target = bank_data[['y']]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.3, random_state = 0)

reg = LogisticRegression(C=1.0, solver = 'newton-cg' )

reg.fit(X_train, y_train)

y_pred = reg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(reg.score(X_test, y_test)))

confusion_matrix = confusion_matrix(y_test, y_pred)
print('confusion_matrix of LOgit:',confusion_matrix)

from sklearn.metrics import classification_report

# classification_report of logit classification
print('classification_report of logit:',classification_report(y_test, y_pred))
```

#### `11702+347 columns are predicted correct, while 267+1248 are predicted wrong`

###  `here the calssification report shows above 90% of people are interested for term deposit`

```python

# ROC Curve

from sklearn.metrics import roc_auc_score, roc_curve


# predict probabilities
probs = reg.predict_proba(X_test)

# keep probabilities for the positive outcome only
probs = probs[:, 1]

auc = roc_auc_score(y_test, probs)
print('AUC - Test Set: %.2f%%' % (auc*100))

# calculate roc curve
fpr, tpr, thresholds = roc_curve(y_test, probs)

# plot no skill
plt.plot([0, 1], [0, 1], linestyle='-')

# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
plt.xlabel('False positive rate')
plt.ylabel('Sensitivity/ Recall')
plt.show()
```

the AUC test score is 86%, which is a good score by the classifier

```python
from sklearn import linear_model, decomposition
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
import numpy as np

from sklearn.preprocessing import LabelEncoder

cols_to_encode = ['job', 'marital', 'education', 'default', 'housing',
                   'loan', 'contact', 'month','poutcome','y']

le = LabelEncoder()

for x in cols_to_encode:
    bank_data[x] = le.fit_transform(bank_data[x])

features = bank_data[['age', 'job', 'marital', 'education', 'default', 'balance', 'housing',
       'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays',
       'previous', 'poutcome']]

target = bank_data[['y']]

X = features
y= target

std_slc = StandardScaler()

pca = decomposition.PCA()

# HYPERPARAMETER TUNING 

logistic_Reg = linear_model.LogisticRegression(solver = 'newton-cg')

pipe = Pipeline(steps= [('std_slc', std_slc), ('pca', pca),('logistic_Reg',logistic_Reg)])

n_components = list(range(1, X.shape[1]+1,1))

C= np.logspace(-4,4,50)

penalty = ['l1','l2','elasticnet']

parameters =  dict(pca__n_components = n_components, logistic_Reg__C = C, logistic_Reg__penalty = penalty)

clf_GS = GridSearchCV(pipe, parameters)

clf_GS.fit(X,y)
```

```python
print('Best penalty:', clf_GS.best_estimator_.get_params()['logistic_Reg__penalty'])

print('Best_C:',clf_GS.best_estimator_.get_params()['logistic_Reg__C'])

print('Best Number of Components:',clf_GS.best_estimator_.get_params()['pca__n_components'])

print(); print(clf_GS.best_estimator_.get_params()['logistic_Reg'])
```
